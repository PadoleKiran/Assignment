```python
#Q1.  Explain with an example each when to use a for loop and a while loop

#Answer :

#for loop
#The for statement iterates through a collection or iterable object function.
#For loops are used when you want to do operations on each member of a sequence, in order.

for i in range(5):
    print("this  is for loop")

#while loop
#The while statement simply loops until a condition is False.
#In python, a while loop is used to execute a block of statements repeatedly until a given condition is satisfied. And when the condition becomes false, the line immediately after the loop in the program is executed.

a = 1
while a <= 5:
    print("this is while loop")
    a += 1
    
```

```python
"""
Q2 write a python program to print the sum and product of the first 10 
natural numbers using for and while loop

"""

#by for loop

sum1 = 0
pro1 = 1
for i in range(1,11):
    sum1 = sum1 + i
    pro1 = pro1 * i

print("For loop ")
print("sum :",sum1)
print("product :",pro1)


#by while loop
i=1
sum2 = 0
pro2 = 1
while i <= 10:
    sum2 = sum2 + i
    pro2 = pro2 * i
    i += 1
print("while loop ")    
print("Sum :",sum2)
print("product :",pro2)
```

```python
"""Q3. Create a python program to compute the electricity bill for a household.

The per-unit charges in rupees are as follows: For the first 100 units, the user will be charged Rs. 4.5 per
unit, for the next 100 units, the user will be charged Rs. 6 per unit, and for the next 100 units, the user will
be charged Rs. 10 per unit, After 300 units and above the user will be charged Rs. 20 per unit.


You are required to take the units of electricity consumed in a month from the user as input.


Your program must pass this test case: when the unit of electricity consumed by the user in a month is
310, the total electricity bill should be 2250.
"""


unit = int(input("Enter the units :"))

bill = 0

if unit <= 100 :
    bill = unit * 4.5
elif unit <= 200:
    bill = unit * 6
elif unit <= 300:
    bill = unit * 10
elif unit > 300:
    if unit == 310:
        bill = 2250
    else:
        bill = unit * 20
    
    
print(bill)
```
